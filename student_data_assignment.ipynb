```python
#Seggregate data in categorical and continuous type

cat=[]
con=[]

for i in df.columns:
    if df[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,3))

sns.countplot(df['guardian'])
plt.title('Countplot of guradian')
plt.show()
```

```python
# pie chart for guradian


a=df['guardian'].value_counts()
a

plt.pie(a, labels=a.index, autopct="%0.01f%%")
plt.show()
```
